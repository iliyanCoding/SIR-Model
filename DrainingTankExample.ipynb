```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
def tank(V, t, k):
    dVdt = -k * V
    return dVdt

V0 = 100 # Starting volume of water
k = 0.3 # Draining rate constant

t = np.linspace(0, 50, 100)

V = odeint(tank, V0, t, args=(k,))

plt.plot(t, V, label="Water volume")
plt.xlabel("Time")
plt.ylabel("Volume of water")
plt.title("Draining Tank Example (ODE)")
plt.legend()
plt.show()
```
